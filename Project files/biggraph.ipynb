```python
import networkx as nx
import matplotlib.pyplot as plt
import random
```

```python
def make_graph(num_nodes, prob, name):
    
    g=nx.fast_gnp_random_graph(num_nodes, prob, seed=None, directed=True)

    weight_dict={}
    weight_lst = list(range(5,50))
    for i in g.nodes:
        weight_dict[i]=random.choice(weight_lst)

    #weight_dict[num_nodes-1]=0 #sink
    nx.set_node_attributes(g, weight_dict, "weight")
    #print(g.nodes[nodes-1]['weight'])

    caplen_lst = list(range(1,10))
    cap = {}
    leng = {}
    for i,j in list(g.edges):
        cap[i,j]=random.choice(caplen_lst)
        leng[i,j]=random.choice(caplen_lst)
    nx.set_edge_attributes(g, cap, "capacity")
    nx.set_edge_attributes(g, leng, "length")
    #print(g.edges[0,1])

    nodes = g.nodes
    connected = list(next(nx.weakly_connected_components(g)))
    not_connected = list(nodes-connected)
    print("Initial Connected nodes- ", connected)
    print("\nInitial Not Connected nodes- ", not_connected)

    if not_connected:
        for i in not_connected:
            if i == num_nodes - 1:
                g.add_edges_from([(i-1,i)], capacity=random.choice(caplen_lst), length=random.choice(caplen_lst))
            
            else :
                g.add_edges_from([(i,i+1)], capacity=random.choice(caplen_lst), length=random.choice(caplen_lst))
               # g.add_edges_from([(i-1,i)], capacity=random.choice(caplen_lst), length=random.choice(caplen_lst))
    
   # remove_edges = list(g.edges(num_nodes-1))
    #g.remove_edges_from(remove_edges)

    print("\n", nx.info(g))
    #print("\n ", g.nodes)
    print("\n Edges - ", g.edges)

    connected = list(next(nx.weakly_connected_components(g)))
    not_connected = list(nodes-connected)
    print("\nFinal Connected nodes - ", connected)
    print("\nFinal Not Connected nodes - ", not_connected)

    #pos = nx.spring_layout(g, seed=9)  # positions for all nodes - seed for reproducibility

    pos = nx.spring_layout(g)

    # nodes
    nx.draw_networkx_nodes(g, pos, node_size=700)

    # edges
    nx.draw_networkx_edges(g, pos, width=6)

    labeldict={}
    for i in g.nodes:
        labeldict[i]=i

    nx.draw_networkx_labels(g,pos, labels=labeldict,font_size=10, font_family="sans-serif")

    # edge weight labels
    nx.draw_networkx_edge_labels(g, pos, g.edges)

    #nx.draw(g)
    plt.show()
    nx.write_gml(g, name)
```

```python
num_nodes = int(input("Enter number of nodes - "))
prob = float(input("Enter probability - "))
name = str(input("Enter name for the graph - ")) + ".gml"
make_graph(num_nodes, prob, name)
```
