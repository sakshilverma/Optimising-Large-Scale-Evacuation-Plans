```python
import networkx as nx
import matplotlib.pyplot as plt
import random
```

```python
g=nx.DiGraph()
g.add_node('A', weight=20)
g.add_node('B', weight=3)
g.add_node('C', weight=5)
g.add_node('D', weight=10)
g.add_node('E', weight=12)
g.add_node('F', weight=17)
g.add_node('G', weight=2)
g.add_node('H', weight=4)
g.add_node('S', weight=0)
g.add_edges_from([('A','B')], capacity=3,length=4)
g.add_edges_from([('A','D')], capacity=1, length=5)            
g.add_edges_from([('A','F')], capacity=4,length=2)               
g.add_edges_from([('B','D')], capacity=1,length=2)                   
g.add_edges_from([('B', 'C')], capacity=2,length=2)
g.add_edges_from([('C', 'E')], capacity=4,length=1)
g.add_edges_from([('C', 'S')], capacity=1,length=3)
g.add_edges_from([('D', 'F')], capacity=5,length=1)
g.add_edges_from([('D', 'E')], capacity=5,length=2)
g.add_edges_from([('E', 'G')], capacity=8,length=3)
g.add_edges_from([('E', 'S')], capacity=3,length=7)
g.add_edges_from([('F', 'G')], capacity=6,length=4)
g.add_edges_from([('G', 'H')], capacity=4,length=4)
g.add_edges_from([('H', 'S')], capacity=2,length=6)
print(nx.info(g))
print(g.edges)
pos = nx.spring_layout(g, seed=9)  # positions for all nodes - seed for reproducibility

# nodes
nx.draw_networkx_nodes(g, pos, node_size=700)

# edges
nx.draw_networkx_edges(g, pos, width=6)

# node labels
labeldict={}
labeldict['A']=20
labeldict['B']=3
labeldict['C']=5
labeldict['D']=10
labeldict['E']=12
labeldict['F']=17
labeldict['G']=2
labeldict['H']=4
labeldict['S']=0
nx.draw_networkx_labels(g,pos, labels=labeldict,font_size=10, font_family="sans-serif")
#nx.draw(G, labels=labeldict, with_labels = True)

# edge weight labels
nx.draw_networkx_edge_labels(g, pos, g.edges)
#edge_labels = nx.get_edge_attributes(g, "capacity")
#nx.draw_networkx_edge_labels(g, pos, edge_labels=edge_labels)

#nx.draw(g)
plt.show()
nx.write_gml(g, "newgraph.gml")
```

```python
G=nx.DiGraph()
lst = list(range(1,10))
G.add_nodes_from([2,3,4,5], weight = random.choice(lst))

for k,v in G.nodes(data=True):
    print(k,v['weight'])
```

```python
g = nx.complete_graph(5)

#weightlist = random.sample(range(5, 50), 5)

weight_dict={}
weight_lst = list(range(5,50))
for i in g.nodes:
    weight_dict[i]=random.choice(weight_lst)
nx.set_node_attributes(g, weight_dict, "weight")
#print(g.nodes[0]['weight'])

caplen_lst = list(range(1,10))
cap = {}
len = {}
for i,j in g.edges:
    cap[i,j]=random.choice(caplen_lst)
    len[i,j]=random.choice(caplen_lst)
nx.set_edge_attributes(g, cap, "capacity")
nx.set_edge_attributes(g, len, "length")
#print(g.edges[0,1])

print(nx.info(g))
print(g.nodes)
print(g.edges)

pos = nx.spring_layout(g, seed=9)  # positions for all nodes - seed for reproducibility

# nodes
nx.draw_networkx_nodes(g, pos, node_size=700)

# edges
nx.draw_networkx_edges(g, pos, width=6)

labeldict={}
for i in g.nodes:
    labeldict[i]=g.nodes[i]['weight']
    
nx.draw_networkx_labels(g,pos, labels=labeldict,font_size=10, font_family="sans-serif")

# edge weight labels
nx.draw_networkx_edge_labels(g, pos, g.edges)

#nx.draw(g)
plt.show()

nx.write_gml(g, "biggraph.gml")
```
