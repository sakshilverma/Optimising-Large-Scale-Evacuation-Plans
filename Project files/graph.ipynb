```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
g=nx.DiGraph()
g.add_node('A', weight=5)
g.add_node('B', weight=7)
g.add_node('C', weight=0)
g.add_node('O', weight=0)
g.add_edges_from([('A','B')], capacity=1,length=3)
g.add_edges_from([('A','C')], capacity=2,length=2)            
g.add_edges_from([('B','O')], capacity=2,length=4)               
g.add_edges_from([('B','C')], capacity=4,length=1)  
#g.add_edges_from([('C','O')], capacity=1,length=5) 
print(nx.info(g))
print(g.edges)
pos = nx.spring_layout(g, seed=5)  # positions for all nodes - seed for reproducibility

# nodes
nx.draw_networkx_nodes(g, pos, node_size=700)

# edges
nx.draw_networkx_edges(g, pos, width=6)

# node labels
labeldict={}
labeldict['A']=5
labeldict['B']=7
labeldict['C']=0
labeldict['O']=0
nx.draw_networkx_labels(g,pos, labels=labeldict,font_size=20, font_family="sans-serif")
#nx.draw(G, labels=labeldict, with_labels = True)

# edge weight labels
nx.draw_networkx_edge_labels(g, pos, g.edges)
#edge_labels = nx.get_edge_attributes(g, "capacity")
#nx.draw_networkx_edge_labels(g, pos, edge_labels=edge_labels)

#nx.draw(g)
plt.show()

nx.write_gml(g, "graph.gml")
```
